```python
import requests

from bs4 import BeautifulSoup

import pandas as pd

url="https://en.wikipedia.org/wiki/List_of_active_Indian_military_aircraft"
```

```python
response=requests.get(url)
```

```python
response.content
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
table = soup.find('table', class_='wikitable')
```

```python
table
```

```python
for i in table:
    print(i)
    print("**********************")
```

```python
table.find_all('tr')[2:]
```

```python
for i in table.find_all('tr')[2:]:
    print(i)
    print("**********************************")
```

```python
aircrafts= {}
for row in table.find_all('tr')[2:]:
    desc =[]
    key = row.find_all("td")[0].text.replace("\n","")
    for i in row.find_all("td")[1:]:
        desc.append(i.text.replace("\n",""))
        
    aircrafts[key]=desc
```

```python
aircrafts
```

```python
Aircrafts ={}
for row in table.find_all('tr')[2:]:
    try:
        key = row.find_all("td")[0].text.strip()
        desc =[]
        
        for i in row.find_all("td")[1:]:
            desc.append(i.text.strip())
        
        Aircrafts[key] = desc
    except IndexError:
        continue
```

```python
Aircrafts
```

```python
import csv 
```

```python
field_names = ['Aircraft', 'Origin', 'Type','Variant','In service','Notes'] 

rows = []
for aircraft, data in Aircrafts.items():
    rows.append(dict(zip(field_names, [aircraft] + data)))
    
with open('aircraft.csv', 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=field_names)
    writer.writeheader()
    writer.writerows(rows)
```
